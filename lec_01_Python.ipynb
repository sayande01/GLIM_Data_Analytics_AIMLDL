### Visualization

```python
df1 = df.groupby('Campaign Name')['Impressions'].mean().sort_values(ascending = False).head(7)
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x = df1.index, y =df1.values, hue = df1.index, palette = 'bright')
plt.xticks(rotation = 45)
plt.show()
```

```python
df2 = df.groupby('Campaign Name')['Cost'].sum().sort_values(ascending = False).head(7)
```

```python
df2.plot(kind = 'pie', autopct = '%1.2f%%')
plt.ylabel(None)
plt.show()
```

```python
sns.histplot(x = 'Impressions', data = df, kde = True)
plt.show()
```
