### Data Filtering

```python
df[df['Location'] == 'Texas']
```

```python
df[df['Average Room Rate'] > 200]
```

```python
#specific columns
df[df['Average Room Rate'] > 200][['Hotel Name', 'Location', 'Average Room Rate']]
```

```python
# dealing with multiple conditions
df[(df['Average Room Rate'] > 200) & (df['Number of Rooms'] > 300) & (df['Location'] == 'Hawaii')]
```

```python
#OR operator
df[(df['Average Room Rate'] > 150) & ((df['Location'] == 'Texas') | (df['Location'] == 'Florida'))]
```
